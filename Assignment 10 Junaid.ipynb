```python
import sqlite3 as sql
import pandas as pd
```

```python
conn = sql.connect('factbook.db')
conn
```

```python
def sql_query(q):
    return pd.read_sql_query(q,conn)
```

```python
q  = '''select * from sqlite_master '''
sql_query(q)
```

## Q# 1
##### List all cities where the area is greater than 5000, along with their population.

```python
q  = '''SELECT c.name,f.area,c.population from cities c
INNER JOIN facts f 
ON 
c.facts_id = f.id
WHERE f.area > 5000
 '''
sql_query(q)
```

### Q # 2
##### List the names of capital cities where the death rate is greater than 2% and the birth rate is less than 20%

```python
q  = '''SELECT c.name from cities c
INNER JOIN facts f 
ON 
c.facts_id = f.id
WHERE c.capital = 1   AND f.death_rate > 2   AND f.birth_rate < 20

 '''
sql_query(q)
```

### Q # 3
##### List the cities with a population growth rate between 1% and 5%, sorted in desc by population growth.

```python
q  = '''select c.name, f.population_growth from cities c
inner join facts f ON c.facts_id = f.id
WHERE f.population_growth BETWEEN 1 AND 5
ORDER BY f.population_growth DESC


 '''
sql_query(q)
```

### Q # 4
##### Find countries that start with &quot;A&quot; and do not have any cities listed in the cities table. The query should return the country name and city name. Using left join

```python
q  = '''select f.name as country_name, c.name as city_name
FROM facts f
Left join cities c 
ON f.id = c.facts_id
WHERE f.name LIKE 'A%' AND c.name IS NULL

 '''
sql_query(q)
```

### Q # 5
##### List all capital cities and their corresponding death rate using a LEFT JOIN.

```python
q  = '''select c.name as capital_city, f.death_rate from cities c
LEFT JOIN facts f 
ON c.facts_id = f.id
WHERE c.capital = 1
 '''
sql_query(q)
```
