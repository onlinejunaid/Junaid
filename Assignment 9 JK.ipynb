```python
import sqlite3 as sql
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sql.connect('database.sqlite')
conn
```

```python
pd.read_sql_query("select * from sqlite_master where type = 'table'",conn)
```

```python
def sql_query(q):
    return pd.read_sql_query(q,conn)
```

```python
q = "select * from Salaries"
sql_query(q)
```

1. What is the average TotalPayBenefits for each JobTitle? exclude ('Not Provided', 'Not provided')

```python
q = '''select JobTitle, avg(TotalPayBenefits) from salaries
    where JobTitle != 'Not Provided' or 'Not provided'
    group by JobTitle'''
sql_query(q)
```

2. Which JobTitle has the most employees in each year? exclude ('Not Provided', 'Not provided')

```python
q = '''select year, JobTitle, count(employeename) from salaries
    where JobTitle != 'Not Provided' or 'Not provided'
    group by year , JobTitle
    order by 3 desc
    limit 10
    '''
sql_query(q)
```

3. What is the total number of unique employees? exclude ('Not Provided', 'Not provided')

```python
q = '''SELECT COUNT(DISTINCT EmployeeName) AS TotalUniqueEmployees FROM salaries
WHERE JobTitle != 'Not Provided' or JobTitle!='Not provided'
'''
sql_query(q)
```

4. Show All Employees Ordered By Their TotalPayBenefits In Descending Order? exclude ('Not Provided', 'Not provided')

```python
q = '''SELECT EmployeeName, TotalPaybenefits FROM salaries
WHERE JobTitle != 'Not Provided' or 'Not provided'
order by 2 desc
'''
sql_query(q)
```
